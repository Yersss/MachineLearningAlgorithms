# DL models in Keras

```python
# importing
import pandas as pd
import numpy as np
from keras.layers import Dense
from keras.models import Sequential
```

```python
dataset = pd.read_csv('hourly_wages.csv')
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
predictors = dataset.iloc[:, 1:]
target = dataset['wage_per_hour']
```

```python
predictors.columns.size
```

```python
n_cols = predictors.columns.size
```

```python
model = Sequential()
```

```python
model.add(Dense(50, activation='relu', input_shape=(n_cols,)))
```

```python
model.add(Dense(32, activation='relu'))
```

```python
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(predictors, target, epochs=10)
```

```python
y_preds = model.predict(predictors)
```

```python
for i,j in zip(target,y_preds):
    print('true ', i, ' pred ', j, ' diff ', i-j)
```
